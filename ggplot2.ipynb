```r
data("faithful")
library(ggplot2)
```

```r
head(faithful)
```

```r
str(faithful)
```

```r
ggplot(data = faithful, mapping = aes(x = eruptions, y = waiting)) +
    geom_point() + 
    geom_smooth(method = "lm")
```

```r
ggplot(data = faithful) + 
    geom_point(aes(x = eruptions, y = waiting, colour = eruptions < 3))
```

```r
ggplot(data = faithful) + 
    geom_point(aes(x = eruptions, y = waiting), colour = "steelblue")
```

```r
ggplot(data = faithful) + 
    geom_histogram(aes(x = eruptions, fill = eruptions < 3.1))
```

```r
ggplot(faithful, aes(x = eruptions, y = waiting)) + 
  geom_density_2d() + 
  geom_point()
```

```r
data("mpg")
head(mpg)
```

```r
table(mpg$class)
```

```r
ggplot(mpg) + 
    geom_bar(aes(x = class))
```

```r
ggplot(mpg) + 
    geom_density(aes(x = hwy))
```

```r
ggplot(mpg) + 
    geom_point(aes(x = displ, y = hwy, colour = class)) +
    scale_colour_brewer(palette = "Set1")
```

```r
ggplot(mpg) + 
  geom_point(aes(x = displ, y = hwy)) + 
  scale_x_continuous(breaks = c(3, 5, 6)) + 
  scale_y_continuous(trans = 'log10')
```

```r
ggplot(mpg) + 
    geom_point(aes(x = displ, y = hwy)) + 
    facet_wrap(~class)
```

```r
ggplot(mpg) + 
  geom_point(aes(x = displ, y = hwy)) + 
  facet_grid(year ~ drv)
```

```r
ggplot(mpg) + 
  geom_bar(aes(x = class)) + 
  coord_polar()
```

```r
ggplot(mpg) + 
  geom_bar(aes(x = class)) + 
  coord_cartesian(ylim = c(0, 40))
```
